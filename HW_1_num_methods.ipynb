```python
# Задача 1.1.21
narr = [10,10**2,10**3,10**4,10**5]
s = [0]*len(narr)
dif=[0]*len(narr)
s_analit=1.0
for j in range(len(narr)):
    for i in range(narr[j]+1):
        s[j] += 24/((i**2+8*i+15)*7)
    dif[j] = s_analit-s[j]
dig=[]
for i in range(len(narr)):
    j=1
    k=0
    while j < 100:
        if dif[i]<=10**(-j):
            k+=1
            j+=1
        else:
            dig.append(k)
            break
print()
for i in range(len(narr)):
    print('Частичная сумма ряда при N =',narr[i],'равна', s[i])
print()
for i in range(len(narr)):
    print('Величина абсолютной погрешности при N =',narr[i],'равна', '{:0.15f}'.format(dif[i]))
print()
for i in range(len(narr)):
    print('Количество верных цифр при N =',narr[i],'равна',dig[i])
print()

import matplotlib.pyplot as plt
plt.bar([str(x) for x in narr],dif)
plt.xlabel('N')
plt.ylabel('Абсолютная погрешность')
plt.title('Абсолютная погрешность при различных N')
plt.show()
plt.bar([str(x) for x in narr],dig)
plt.xlabel('N')
plt.ylabel('Количество верных цифр')
plt.title('Количество верных цифр при различных N')
plt.show()
```

```python
#Задание 1.9.3
import numpy as np
alpha = 0.05
beta = 0.1
A = np.array([[3,1,13],[13.4,11.4,23],[5,3,15]])
detA = np.linalg.det(A)
print()
print('Определитель исходной матрицы равен', detA)
if detA != 0:
    print('Матрица A обратима, так как определитель не равен 0')
else: 
    print('Матрица A необратима так как определитель равен 0')
print()
print()
def solution(er):
    B = A.copy()
    result = []
    error = [-er,er]
    for a_11 in error:
        B[0][0] = A[0][0] + a_11
        for a_12 in error:
            B[0][1] = A[0][1] + a_12
            for a_13 in error:
                B[0][2] = A[0][2] + a_13
                for a_21 in error:
                    B[1][0] = A[1][0] + a_21
                    for a_22 in error:
                        B[1][1] = A[1][1] + a_22
                        for a_23 in error:
                            B[1][2] = A[1][2] + a_23
                            for a_31 in error:
                                B[2][0] = A[2][0] + a_31
                                for a_32 in error:
                                    B[2][1] = A[2][1] + a_32
                                    for a_33 in error:
                                        B[2][2] = A[2][2] + a_33
                                        result.append(np.linalg.det(B))
    print('Минимальный определитель матрицы A с погрешностью', error[1],'равен', min(result))
    print('Максимальный определитель матрицы A с погрешностью',error[1],'равен', max(result))
    print()
    if min(result)*max(result)<0:
        print('Матрица А с погрешностью', error[1],'не обратима')
    else:
        print('Матрица А с погрешностью', error[1],'обратима') 
    print()    
    print('Относительная погрешность матрицы А с внесением погрешности',error[1],
          'равна', abs(detA - np.linalg.det(B))/abs(detA))
    print()
    print()
solution(alpha)
solution(beta)

```
